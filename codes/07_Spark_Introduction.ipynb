```python
from pyspark import SparkContext
sc = SparkContext.getOrCreate()
sayilar = [12,43,3,2,51,42,59,14]

kareler = []
for sayi in sayilar:
  kare = sayi*sayi
  kareler.append(kare)
print(kareler)

sayilarRDD = sc.parallelize(sayilar) # Bir python listesini RDD (Resilient Distributed Dataset) ye dönüştürüyoruz

kareleRDD = sayilarRDD.map(lambda x : x * x) # Verilen işlemi bütün elemanlara uygular
print(kareleRDD.collect())
```

```python
pointsRDD = sc.textFile("points.txt")
def splitAndConvert(satir):
  arr = satir.split(',')
  x = int( arr[0] )
  y = int( arr[1] )
  return (x,y)
pointsRDD = pointsRDD.map(lambda line : splitAndConvert(line))
pointsRDD = pointsRDD.map(lambda a : (a[0],a[1],a[0]+a[1]) )
pointsRDD = pointsRDD.sortBy(lambda b : b[2], ascending=False )
pointsRDD = pointsRDD.filter(lambda c : c[2]<20)
pointsRDD.collect()
```

```python
#RDD  -->  Transformation (map,sortBy, filter, .... ) , Action (collect, take, reduce )
# RDD -> T  --> RDD
# RDD -> A --> non-RDD ( int, list ..vs )
rdd1 = sc.parallelize([[12,23,34],[9,4,3,5], [23,4],[ 3 , 43 ,34 ,56], [45],[12]])
rdd1 = rdd1.flatMap(lambda x : x)
rdd1.collect()
```

```python
textRDD = sc.textFile('text.txt')
textRDD = textRDD.flatMap(lambda line : line.split(' '))
textRDD = textRDD.filter(lambda kelime : len(kelime)>0)
textRDD.collect()
```
