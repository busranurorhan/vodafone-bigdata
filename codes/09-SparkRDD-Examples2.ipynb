```python
from pyspark import SparkContext
sc = SparkContext.getOrCreate()
lotteryRDD = sc.textFile("datasets/Lottery.txt")

def splitAndTake(line):
    arr = line.split("rakamlarNumaraSirasi")
    abc= []
    for a in arr:
        b = a[4:30]
        c = b.split(" - ")
        if (len(c) > 4):
            abc.append(c)
    return abc

lotteryRDD = lotteryRDD.flatMap(lambda x : splitAndTake(x))
lotteryRDD.collect()
```
