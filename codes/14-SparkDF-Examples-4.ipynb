```python
from pyspark.sql import SparkSession

spark = SparkSession.Builder().appName('Spark DataFrame Introduction').getOrCreate()
actorsDF = spark.read.option('inferSchema','true')\
                        .option('header','true')\
                        .option('delimiter',',').csv('datasets/join-actors.txt')

seriesDF = spark.read.option('inferSchema','true')\
                        .option('header','true')\
                        .option('delimiter',',').csv('datasets/join-series.txt')
seriesDF.show()
actorsDF.show()

#Without SQL
joinedDF = actorsDF.join(seriesDF,'movieid','outer') #left , right ,inner ,outer
joinedDF.show(40)

#With SQL
actorsDF.createOrReplaceTempView('actors')
seriesDF.createOrReplaceTempView('series')
#sqlDF = spark.sql('select a.*,s.* from actors a,series s where s.movieid=a.movieid')
sqlDF = spark.sql('select a.*,s.* from actors a full outer join series s on s.movieid=a.movieid')
sqlDF.show(40)
```
